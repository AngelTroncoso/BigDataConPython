```python
from datetime import timedelta

start = time.monotonic()
#nuestro codigo
time.sleep(1)
end = time.monotonic()

print(timedelta(seconds=end-start))
```

```python
x = df["PrecioMax"]
```

```python
Q1 = np.percentile(x,25)
Q3 = np.percentile(x,75)
rangointer = Q3 -Q1
```

```python
umbralsuperior = Q3 + 1.5*rangointer
umbralinferior = Q1 - 1.5*rangointer
```

```python
umbralinferior
```

```python
umbralsuperior
```

```python
np.mean(x>umbralsuperior)
```

```python
np.mean(x<umbralinferior)
```

```python
from sklearn.covariance import EllipticEnvelope
```

```python
outliers = EllipticEnvelope(contamination = .01)
```

```python
var_list = ["PrecioMin","PrecioMax","MPG_Ciudad","MPG_Carretera","RPM","RevolucionesPorMilla","CapacidadTanqueCombustible","Pasajeros"]
```

```python
x = np.array(df.loc[:,var_list].dropna())
```

```python
outliers.fit(x)
```

```python
pred = outliers.predict(x)
```

```python
pred
```

```python
elips_outliers = np.where(pred==-1)[0]
```

```python
elips_outliers
```
