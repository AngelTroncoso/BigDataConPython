```python
import time

start = time.time()
#nuestro codigo
time.sleep(0.5)

end =  time.time()

print(end - start)
```

```python
from datetime import timedelta

start = time.monotonic()
#nuestro codigo
time.sleep(1)
end = time.monotonic()

print(timedelta(seconds=end-start))
```

```python
start = time.time()
#nuestro codigo
time.sleep(0.5)

end =  time.time()

start_2 = time.time()
#nuestro codigo
time.sleep(0.501)

end2 =  time.time()

print(end - start > end2 - start_2)
```

```python
# prompt: traducir columnas al español

import pandas as pd

# Spanish translations for column names
column_translations = {
    'Manufacturer': 'Fabricante',
    'Model': 'Modelo',
    'Type': 'Tipo',
    'Min.Price': 'PrecioMin',
    'Price': 'Precio',
    'Max.Price': 'PrecioMax',
    'MPG.city': 'MPG_Ciudad',
    'MPG.highway': 'MPG_Carretera',
    'AirBags': 'AirBags',
    'DriveTrain': 'Traccion',
    'Cylinders': 'Cilindros',
    'EngineSize': 'TamañoMotor',
    'Horsepower': 'CaballosFuerza',
    'RPM': 'RPM',
    'Rev.per.mile': 'RevolucionesPorMilla',
    'Man.trans.avail': 'TransmManualDisp',
    'Fuel.tank.capacity': 'CapacidadTanqueCombustible',
    'Passengers': 'Pasajeros',
    'Length': 'Longitud',
    'Wheelbase': 'DistanciaEntreEjes',
    'Width': 'Anchura',
    'Turn.circle': 'CircunferenciaGiro',
    'Rear.seat.room': 'EspacioAsientosTraseros',
    'Luggage.room': 'EspacioEquipaje',
    'Weight': 'Peso',
    'Origin': 'Origen',
    'Make': 'Marca'
}

# Rename columns using the translations
df = df.rename(columns=column_translations)

# Display the updated DataFrame with Spanish column names
df.head()
```
